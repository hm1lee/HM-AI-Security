## 3. 사이트 크롤링

```python
linkList = set()
import re

host = "http://www.naver.com"
html = urlopen(host)
bsObj = BeautifulSoup(html, "html.parser")

a = bsObj.findAll("a", href=re.compile("^http?://[a-z..]*naver\.com"))

for link in a:
    linkList.add(link["href"])
    
set(linkList)
```

```python
# /~ 형태 조사

linkList = set()

a = bsObj.findAll("a", href=re.compile("^\/"))

for link in a:
    linkList.add(host[:-1]+ link["href"])
    
linkList
```
