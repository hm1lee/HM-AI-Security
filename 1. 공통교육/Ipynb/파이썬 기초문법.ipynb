#### Tuple (서로 다른 형식의 데이터를 집합으로 생성 가능)

* 튜플의 요소는 변조하거나 삭제가 불가능하다. 
* 튜플에는 가변하는 리스트를 포함할 수 있다.
* 서로 다른 형식의 데이터를 집합에 포함시킬 수 있다.
* 생성된 튜플에 데이터를 추가할 수 없다.

```python
tuple1 = 1234, 5678, 'abcd', 'efgs'
tuple1
```

```python
tuple2 = tuple1, (1,2,3,4,5)
tuple2
```

#### Set (중복된 요소가 없는 정렬되지 않은 집합)

* 멤버를 검사하고 중복 항목을 제거
* Union, intersection, diffrence와 같은 수학적 연산을 지원
* 중괄호나 set() 함수를 사용해서 생성

```python
market = {'beef', 'pork', 'lamb', 'beef', 'lamb', 'pork'}
```

```python
market ##데이터 중복 확인해보기
```

```python
'beef' in market
```

```python
'perl' in market
```

```python
a = set('12345')
b = set('45676')
```

```python
a
```

```python
a-b ## 차집합
```

```python
a|b ## 힙집합
```

```python
a&b ## 교집합
```

```python
a^b ## Or
```

```python
## 1번 구현하기

a = set('aabbachdhefgg')
b = set('hdhefokkillmm')
c = set('efggjnnjpip')
```

```python
# a와 b의 합집합
a|b
```

```python
# a와 c의 차집합
a-c
```

```python
# a와 b,c의 교집합
a&b&c
```

#### Dictionary
* "key"를 사용하여 "value"를 인덱스
* 키는 strings나 number가 될 수 있으나, 변하지 않아야 함
* 키는 고유한 성질을 가져야 한다.

```python
student = {'Sun':1, 'Choi':2, 'll':3}
student['hi'] = 0
student
```

```python
student['Sun']
```

```python
del student['Choi']
student
```

```python
sorted(student.keys())
```

```python
'Sun' in student
```

```python
'll' not in student
```
