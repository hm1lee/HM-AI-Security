### 데이터 시각화 (matplotlib)

01. 데이터 시각화란?
- 특이값을 찾아내거나 데이터 변형이 필요한지 알아보거나 모델에 대한 아이디어를 찾기 위한 과정
- matplotlib : 주로 2D 도표를 위한 패키지, Ipython과 GUI 툴킷과 함께 통합하여 컴퓨팅을 위한 다양한 기능을 겸비, 저수준의 그래프 그리기 가능

2. ndarray 그래프 그리기 
- plot, close 같은 matplotlib API 함수는 matplotlib.pyplot 모듈에 포함되어 있음
- plt라는 이름으로 임포트 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a1 = np.arange(10)
a1
```

```python
plt.plot(a1)
plt.show()
```

```python
# x축은 index, y축은 값임을 알 수 있음

pd.Series(a1)
```

```python
# 랜덤한 ndarray 값 생성

a = np.random.randn(10)
a
```

```python
plt.plot(a)
plt.show()
```

```python
def f1(x) :
    return 2*x + 3


arr1 = np.arange(-10, 10, 0.1)
arr1
```

```python
arr2= f1(arr1)
plt.plot(arr1, arr2)
plt.show()
```

3. 그래프 꾸미기
- matplotlib에서 plot을 사용하여 그래프를 다양하게 꾸밀 수 있음
- 색상, 마커, 라인스타일, 드로우스타일, 축 이름, 타이틀 등 다양한 꾸미기 가능

```python
arr1 = np.random.randn(10)
plt.plot(arr1)

arr2 = np.random.randn(10)
plt.plot(arr2)

plt.show()
```

```python
# 생상 지정하여 값 전달 가능

plt.plot(arr1, color='r')
plt.plot(arr2, color='b')
plt.show()
```
